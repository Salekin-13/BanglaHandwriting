```python
from typing import List, Dict, Any
import json

def _get_line_strings_from_json_file(filename: str) -> List[str]:
    """Get the text content of each line."""
    json_line_elements = _get_line_elements_from_json_file(filename)
    return [_get_text_from_json_element(el) for el in json_line_elements]

def _get_text_from_json_element(line_element: Dict[str, Any]) -> str:
    """Extract the text content from a JSON line element."""
    if "label" in line_element:
        return line_element["label"]
    return ""

def _get_line_elements_from_json_file(filename: str) -> List[Dict[str, Any]]:
    with open(filename, 'r', encoding='utf-8') as file:
        json_data = file.read()

        # Parse JSON data
        parsed_data = json.loads(json_data)

        # Extract the "line" elements
        lines = parsed_data["line"] if "line" in parsed_data else []

        return lines

# Example usage
filename = "dataset/raw_images/B_IAM/json/0_21_0.json"
line_strings = _get_line_strings_from_json_file(filename)
print(line_strings)
```

```python
from typing import List, Dict, Any
import json
from pathlib import Path

EXTRACTED_DATASET_DIRNAME = "dataset/raw_images/B_IAM"  # Update with the actual path

class YourClass:
    def json_filenames(self):
        """A list of the filenames of all .json files, which contain label information."""
        return list((Path(EXTRACTED_DATASET_DIRNAME) / "json").glob("*.json"))
    
    def paragraph_string_by_id(self):
        """A dict mapping a BIAM form id to its paragraph text."""
        return {id_: "\n".join(line_strings) for id_, line_strings in self.line_strings_by_id.items()}
    
    @property
    def line_strings_by_id(self):
        """A dict mapping a BIAM form id to its list of line texts."""
        return {filename.stem: self._get_line_strings_from_json_file(filename) for filename in self.json_filenames()}

    def _get_line_strings_from_json_file(self, filename: Path) -> List[str]:
        """Get the text content of each line."""
        json_line_elements = self._get_line_elements_from_json_file(filename)
        return [self._get_text_from_json_element(el) for el in json_line_elements]

    def _get_text_from_json_element(self, line_element: Dict[str, Any]) -> str:
        """Extract the text content from a JSON line element and replace '&garb;' with ''."""
        if "label" in line_element:
            return line_element["label"]
        return ""

    def _get_line_elements_from_json_file(self, filename: Path) -> List[Dict[str, Any]]:
        with open(filename, 'r', encoding='utf-8') as file:
            json_data = file.read()
            # Parse JSON data
            parsed_data = json.loads(json_data)

            # Extract the "line" elements
            lines = parsed_data["line"] if "line" in parsed_data else []

            return lines

# Create an instance of YourClass
your_instance = YourClass()

# Example usage
paragraphs_by_id = your_instance.paragraph_string_by_id()
for form_id, paragraph in paragraphs_by_id.items():
    print(f"Form ID: {form_id}")
    print(f"Paragraph:\n{paragraph}")
    print("-" * 40)
```

```python
import torch
import re
from typing import Dict, List

def convert_strings_to_labels(strings: List[str], mapping: Dict[str, int], length: int) -> torch.Tensor:
    labels = torch.ones((len(strings), length), dtype=torch.long) * mapping["<P>"]
    for i, string in enumerate(strings):
        string = string.replace("&garb", "<G>")
        string = string.replace("&under", "<U>")
        string = string.replace("&eng", "<ENG>")
        string = string.replace("&deg", "<DEG>")
        
        tokens = re.findall(r"(\s+|<G>|<U>|<ENG>|<DEG>|[^\s<GUE>]+)", string)  # Split the modified string
        
        char_tokens = []
        for token in tokens:
            if token in ["<G>", "<U>", "<ENG>", "<DEG>"]:
                char_tokens.append(token)
            else:
                char_tokens.extend(list(token))  # Split the word into characters
                
        tokens = ["<S>", *char_tokens,"<E>"]  # Add the special start token
        for ii, token in enumerate(tokens):
            labels[i, ii] = mapping.get(token, mapping["<P>"])  # Use mapping["<P>"] if token not in mapping
    return labels
```

```python
import json
import numpy as np
def get_dataset_properties():
    """Return properties describing the overall dataset."""
    with open("dataset/processed_data/B_IAM_paragraphs/_properties.json", "r", encoding="utf-8") as f:
        properties = json.load(f)

    def _get_property_values(key: str) -> list:
        return [_[key] for _ in properties.values()]

    crop_shapes = np.array(_get_property_values("crop_shape"))
    aspect_ratios = crop_shapes[:, 1] / crop_shapes[:, 0]
    return {
        "label_length": {
            "min": min(_get_property_values("label_length")),
            "max": max(_get_property_values("label_length")),
        },
        "num_lines": {"min": min(_get_property_values("num_lines")), "max": max(_get_property_values("num_lines"))},
        "crop_shape": {"min": crop_shapes.min(axis=0), "max": crop_shapes.max(axis=0)},
        "aspect_ratio": {"min": aspect_ratios.min(), "max": aspect_ratios.max()},
    } 
```
