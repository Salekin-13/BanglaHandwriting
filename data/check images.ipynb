```python
import json

# Load the JSON data from the file
with open("dataset\processed_data\B_IAM_paragraphs\_properties.json", "r") as f:
    image_data = json.load(f)

aspect_ratios = []

# Calculate aspect ratios for each image
for iam_id, data in image_data.items():
    width, height = data["crop_shape"]
    aspect_ratio = width / height
    aspect_ratios.append(aspect_ratio)

# Calculate average aspect ratio
average_aspect_ratio = sum(aspect_ratios) / len(aspect_ratios)
print("Average Aspect Ratio:", average_aspect_ratio)
```

```python
from PIL import Image

def resize_image(input_path, output_path, target_size=(1000, 1000)):
    try:
        image = Image.open(input_path)
        resized_image = image.resize(target_size, resample=Image.BILINEAR)
        resized_image.save(output_path)
        print(f"Image resized and saved to {output_path}")
    except Exception as e:
        print(f"Error: {e}")

if __name__ == "__main__":
    input_image_path = image_path  # Replace with the actual path to your input image
    output_image_path = Path(B_iam.EXTRACTED_DATASET_DIRNAME)/ "forms" /"trial_image.jpg"  # Replace with the desired output path

    target_size = (1000, 1000)
    resize_image(input_image_path, output_image_path, target_size)
```

```python
import os
import shutil

# Source folder containing JPG files
source_folder = "D:/Thesis/Main/dataset/raw_images/B_IAM/forms"

# Folder containing the corresponding files to be extracted
corresponding_folder = "D:\Bangla Handwriting\enhanced_images(contrast changed)"

# Target folder to save the extracted files
target_folder = "D:/Thesis/Main/dataset/my_raw"

# Create the target folder if it doesn't exist
os.makedirs(target_folder, exist_ok=True)

# Get the list of JPG file names in the source folder
jpg_file_names = [filename for filename in os.listdir(source_folder) if filename.lower().endswith(".jpg")]

# Loop through the JPG file names and copy the corresponding files to the target folder
for jpg_file_name in jpg_file_names:
    source_file_path = os.path.join(corresponding_folder, jpg_file_name)
    target_file_path = os.path.join(target_folder, jpg_file_name)
    
    # Check if the corresponding file exists and is a file
    if os.path.isfile(source_file_path):
        shutil.copy2(source_file_path, target_file_path)
        print(f"Extracted and saved: {jpg_file_name}")
    else:
        print(f"Corresponding file not found for: {jpg_file_name}")

print("Extraction and saving complete.")
```
